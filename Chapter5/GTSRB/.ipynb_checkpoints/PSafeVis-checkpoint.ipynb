```python
posterior = "VOGN"
bound = "chernoff"

import os
import json
def load(post, bound):
    entries = []
    path = "ExperimentalLogs/%s_%s.log"%(post, bound)
    with open(path) as log_file:
        for line in log_file:
            #print(type(json.loads(line)))
            entries.append(json.loads(line))
    return entries

log_info = load(posterior, bound)
```
